#### Imports

```python
import pandas as pd
subcellular_locals_df = pd.read_csv('subcellular_location.tsv', sep='\t')
```

#### display file

```python
subcellular_locals_df.head()
```

#### melt

```python
subcell_local_melt = pd.melt(subcellular_locals_df, id_vars=['Gene', 'Gene name'], value_vars=['Main location', 'Reliability'])
subcell_local_melt
```

#### pivot

```python
subcell_local_pivot = (
    subcellular_locals_df[['Gene name', 'Main location', 'Additional location', 'Reliability']]
    .drop_duplicates()
    .pivot(index='Gene name', columns='Main location')
)
subcell_local_pivot
```

#### aggregation

```python
#aggregates by 'main location', returning the number of genes with the corresponding main location
subcell_local_agg = (
    subcellular_locals_df[['Main location', 'Gene name']]
    .groupby('Main location')
    .count()
)
subcell_local_agg
```

#### iteration

```python
#iterate over the dataframe (in this case the first 10) and print out the main location
for index,row in subcellular_locals_df[:10].iterrows():
    print(f'The main location for {row["Gene name"]} is: {row["Main location"]}')
```
